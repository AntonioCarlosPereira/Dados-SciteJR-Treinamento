```python
# Converter Release_Date em datetime
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Criar nova coluna apenas com o ano
df['Year'] = df['Release_Date'].dt.year
```

```python
df[['Release_Date', 'Year']].head(10)
```

```python
# Separar os gêneros em múltiplas colunas (One-Hot Encoding)
df_genres = df['Type'].str.get_dummies(sep=', ')

# Concatenar com o df original
df = pd.concat([df, df_genres], axis=1)

df 
```

```python
# Separar países em colunas (cuidado: pode gerar MUITAS colunas)
df_countries = df['Country'].str.get_dummies(sep=', ')

# Concatenar com o df original
df = pd.concat([df, df_countries], axis=1)
df
```

```python
df.info()
```

```python
df['Director'] = df['Director'].fillna('Unknown')
df['Cast'] = df['Cast'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Rating'] = df['Rating'].fillna('Not Rated')
```

```python
# Descarta colunas 'Show_Id' e 'Description'
df = df.drop(['Show_Id', 'Description'], axis=1)
```

```python
df.head()
df.columns
```

```python
df.info()
```

```python
#converter todas as colunas booleanas geradas para True/False
df[df_genres.columns] = df[df_genres.columns].astype(bool)
df[df_countries.columns] = df[df_countries.columns].astype(bool)
df.info()
```

```python
df.info('deep')
```

```python
df['Year']
```

```python
df['Year'] = df['Year'].astype('float16')
```

```python
df.info('deep')
```

```python
# Extrair números da coluna Duration
duration_numbers = df['Duration'].str.extract(r'(\d+)')[0].astype(float)

# Criar colunas separadas
duration_minutes = duration_numbers.where(df['Duration'].str.contains('min')).astype('Int64')
duration_seasons = duration_numbers.where(df['Duration'].str.contains('Season')).astype('Int64')

# Concatenar corretamente apenas uma vez
df = pd.concat([
    df,
    duration_minutes.rename('Duration_Minutes'),
    duration_seasons.rename('Duration_Seasons')
], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionar colunas numéricas que queremos analisar
numerical_cols = ['Year', 'Duration_Minutes', 'Duration_Seasons']

# Copiar os dados
df_numeric = df[numerical_cols].copy()

# --- StandardScaler ---
standard_scaler = StandardScaler()
df_standard_scaled = df_numeric.copy()
df_standard_scaled[numerical_cols] = standard_scaler.fit_transform(df_numeric)

# --- MinMaxScaler ---
minmax_scaler = MinMaxScaler()
df_minmax_scaled = df_numeric.copy()
df_minmax_scaled[numerical_cols] = minmax_scaler.fit_transform(df_numeric)

# --- Boxplot StandardScaler ---
plt.figure(figsize=(10,6))
sns.boxplot(data=df_standard_scaled[numerical_cols])
plt.title('Distribuição das Variáveis Numéricas (StandardScaler)')
plt.ylabel('Valor Escalonado')
plt.show()

# --- Boxplot MinMaxScaler ---
plt.figure(figsize=(10,6))
sns.boxplot(data=df_minmax_scaled[numerical_cols])
plt.title('Distribuição das Variáveis Numéricas (MinMaxScaler)')
plt.ylabel('Valor Escalonado')
plt.show()
```

```python
# Pergunta 9 - Principais países produtores
country_counts = (
    df['Country']
    .dropna()
    .str.split(', ')    # divide casos com múltiplos países
    .explode()          # transforma lista em linhas
    .value_counts()
)

# Remover 'Unknown' se aparecer
country_counts = country_counts[country_counts.index != 'Unknown']

# Top 10
top10_countries = country_counts.head(10)

# Visualização
plt.figure(figsize=(10,6))
sns.barplot(x=top10_countries.values, y=top10_countries.index, palette="viridis")
plt.title("Top 10 países produtores de conteúdo na Netflix")
plt.xlabel("Número de títulos")
plt.ylabel("País")
plt.show()
```
